```python

"""To use PCA Algorithm for dimensionality reduction.
 You have a dataset that includes measurements for diff
 (alcohol, ash, magnesium, and so on). Apply PCA algori
 so that most variations in the measurements of the var
 number of principal components so that it is easier to
 white wine by inspecting these principal components."""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df=pd.read_csv('Wine.csv')

# Separating features and target variable
X = df.drop(columns=['Customer_Segment']) # Features
y = df['Customer_Segment'] # Target variable (Type of

# Scatter plot before applying PCA
plt.figure(figsize=(10, 6))
plt.scatter ( X[y == 1] ['Alcohol'], X[y == 1] ['Malic_Acid'], label='Customer Segment 1')
plt.scatter ( X[y == 2] ['Alcohol'], X[y == 2] ['Malic_Acid'], label='Customer Segment 2')
plt.scatter ( X[y == 3] ['Alcohol'], X[y == 3] ['Malic_Acid'], label='Customer Segment 3')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.legend()
plt.title('Scatter Plot (Original Data)')
plt.show()

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components =2)
X_pca = pca.fit_transform(X_scaled)

# Create a new DataFrame with the first two principal
pca_df = pd.DataFrame (data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Customer_Segment'] = y

# Visualize the data using the first two principal com
plt.figure(figsize =(10,6))
plt.scatter(pca_df [pca_df['Customer_Segment'] ==1]['Principal Component 1'], pca_df [pca_df['Customer_Segment'] ==1]['Principal Component 2'], label='Customer Segment 1')
plt.scatter(pca_df[pca_df['Customer_Segment'] ==2]['Principal Component 1'], pca_df[pca_df['Customer_Segment'] ==2]['Principal Component 2'], label='Customer Segment 2')
plt.scatter(pca_df[pca_df['Customer_Segment'] ==3]['Principal Component 1'], pca_df[pca_df['Customer_Segment'] ==3]['Principal Component 2'], label='Customer Segment 3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA: Wine Dataset')
plt.show()


 
```
